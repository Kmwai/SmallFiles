This Document leads you through the installation of the packing node for the dCache SmallFiles patch.

First we will get all the needed files from the public repository on github. Make sure you have installed <tt>git</tt>!

```python
!git clone https://github.com/kschwank/SmallFiles.git
```

Now we copy the scripts to their appropriate locations:

```python
!cp SmallFiles/src/skel/usr/local/bin/pack-files* /usr/local/bin/!cp SmallFiles/src/skel/usr/local/bin/check-archives.sh /usr/local/bin/
```

Next we will create a mountpoint for the nfs export of the dCache server

```python
!mkdir -p /pnfs/4
```

And then mount the nfs export to that location.Make sure you edit this to match your setup and make sure the server is running!

```python
!mount -o intr,minorversion=1 ceph-mon1:/data /pnfs/4
```

Here you can check, if you see the files you expect in the nfs mount:

```python
!ls -l /pnfs/4
```

Finally, we create an entry in your systems crontab file to periodically start the packing script. The parameters of the pack-files script have the following meanings:1. nfs export directory1. local mount point1. directory for the small files hsm1. target size for the archives1. time period after which remaining files are collected and archived, independent of their combined size

```python
!echo '*  *  *  *  *   root     /usr/local/bin/pack-files-by-group.sh "/data" "/pnfs/4" "hsm" 10000000000 1440 2>&1' >> /etc/crontab
```
