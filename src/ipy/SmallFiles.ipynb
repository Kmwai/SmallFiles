```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

createres = pd.io.parsers.read_csv('create.csv', header=0)
findres = pd.io.parsers.read_csv('find.csv',header=0)
lsres = pd.io.parsers.read_csv('ls.csv', header=0)
statres = pd.io.parsers.read_csv('stat.csv', header=0)
writeres = pd.io.parsers.read_csv('write.csv', header=0)
readres = pd.io.parsers.read_csv('read.csv', header=0)
# pathofres = pd.io.parsers.read_csv('pathof.csv', header=0)                              
# tarres = ps.io.parsers.read_csv('tar.csv', header=0)
# zipres = ps.io.parsers.read_csv('zip.csv', header=0)

results = { 'create' : createres, 'find' : findres, 'ls' : lsres, 'stat' : statres, 'write' : writeres, 'read' : readres }

for op in results.keys():
    df = results[op]
    df['op'] = op
    df = df[(df['file_size'] < 2000000)]
    df['fps'] = df['file_count'] / df['total']
    df['bps'] = df['fps'] * df['file_size']
    results[op] = df
    
```

```python
allres = pd.concat(results.values())
allres.head()
```
